```python
from ase import Atoms
from gpaw import GPAW

d = 0.74
a = 6.0

atoms = Atoms('H2',
              positions=[(0, 0, 0),
                         (0, 0, d)],
              cell=(a, a, a))
atoms.center()

#calc = GPAW(nbands=2,verbose=False, txt='h2.txt')
calc = GPAW(nbands=1, xc='PBE',gpts=(24, 24, 24))
atoms.calc = calc
print(atoms.get_forces())
```
